```python
import json
```

```python
vt_path = "/content/drive/MyDrive/vérité_terrain_ner.json"
```

```python
with open(vt_path, 'r') as f:
    data = json.load(f)
```

Structure des données d'entraînement attendue par spacy : https://spacy.io/api/data-formats#dict-input

```python
training_data = []
for item in data:
    text = item['text']
    entities = []

    for entity in item['label']:
        start = entity['start']
        end = entity['end']
        label = entity['labels'][0]
        entities.append((start, end, label))

    training_data.append((text, {"entities": entities}))
```

```python
training_data
```

```python
import spacy
from spacy.training import Example
```

```python
spacy.cli.download('fr_dep_news_trf')
```

```python
nlp = spacy.load('fr_dep_news_trf')
```

```python
nlp
```

classe `Example` de spacy : https://spacy.io/api/example

```python
examples = []
```

```python
for text, annotations in training_data:
    doc = nlp.make_doc(text)
    example = Example.from_dict(doc, annotations)
    examples.append(example)
```

Méthode `evaluate()` de spacy : https://spacy.io/api/language#evaluate

```python
scores = nlp.evaluate(examples)
```

```python
scores
```

```python
print(scores)
```

- **précision** : vrais positifs, ici c'est donc la proportion d'entités nommées qui sont vraiment correctes. Mesure la
- **rappel** : parmi les entitées nommées présentes dans la vérité terrain, combien ont été détectées positivement par le modèle ?
- **score F1** : moyenne harmonique entre précision et rappel.

```python
print("Précision (ents_p):", scores["ents_p"])
print("Rappel (ents_r):", scores["ents_r"])
print("F1 (ents_f):", scores["ents_f"])
```

```python
print("Détail par type:", scores["ents_per_type"])
```
